```python
# Read this csv file car data.csv and display the basic information like memory and data types for this data frame.
# Write Python Code for the following cases:
# 1.How Many ritz car are there with kms driven more than 30000km.
# 2.How many Petrol cars are of the year 2017 and whose selling price > 10 lakhs. 
# 3.How many swift cars are there with selling price < 4 Lakhs.
# 4.How many AutomaƟc Transmission Petrol Car are of the year 2015 whose selling price is > 10 Lakhs.
# 5.List out Vehicles with AutomaƟc Transmission and selling price < 1 Lakh.
# 6.How Many Petrol Vehicles are there with kms driven more than 50000kms and Year is between 2010 to 2015(both Year 
# included).
# 7.List out the cars whose price difference between present price and selling price is > 15 lakhs.
# 8.How many Petrol vehicles are there whose kms driven < 5000km and selling price < 50000.

import pandas as pd
import numpy as np
df=pd.read_csv('car data.csv')

# INFO
df.info()

# 1
car=df[(df['Kms_Driven']>30000) & (df['Car_Name']=='ritz')]
print(len(car))

# 2
car1=df[(df['Fuel_Type']=='Petrol') & (df['Year']==2017) & (df['Present_Price']>10)]
print(len(car1))

# 3
car2=df[(df['Car_Name']=='swift') & (df['Selling_Price']<4)]
print(len(car2))

# 4
car3=df[(df['Fuel_Type']=='Petrol') & (df['Year']==2015) & (df['Selling_Price']>10) &(df['Transmission']=='Automatic')]
print(len(car3))

# 5
car4=df[(df['Selling_Price']<1) &(df['Transmission']=='Automatic')]
car4

# 6
car5=df[(df['Kms_Driven']>50000) & (df['Year']>=2010) & (df['Year']<=2015) & (df['Fuel_Type']=='Petrol')]
print(len(car5))

# 7
car6=df[(df['Present_Price']-df['Selling_Price'])>15]
print(len(car6))

# 8
car7=df[(df['Kms_Driven']<5000) & (df['Selling_Price']<0.5) &(df['Fuel_Type']=="Petrol")]
print(len(car7))

```
